# Question 1

```

def getStats(array):
  sum = 0
  size = 0
  max = float('-inf')
  min = float('inf')
  
  for number in array:
    size += 1
    sum += number
    if (number > max):
      max = number
    if (number < min):
      min = number
  mean = sum/size
  std = getStd_dev(array, mean, size)
  array.sort()
  p75 = getNth_percentile(75, array, size)
  median = getNth_percentile(50, array, size)
  p25 = getNth_percentile(25, array, size)
  return max, min, mean, std, median, p75, p25
  
def getStd_dev(array, mean, size):
  sum = 0
  for number in array:
    sum += (number-mean) **2
  return (sum/size) ** 0.5
  
def getNth_percentile(n, sorted_array, size):
  L = n/100*size
  L = round(L)
  return sorted_array[L]

a=[]
b=[]
c=[]
d=[]
e=[]
Max=[]
Min=[]
Mean=[]
Stddev=[]
median=[]
p75=[]
p25=[]
stats={'Max':0, 'Min':0, 'Mean':0, 'stddev':0, 'median':0, 'p75':0, 'p25':0}
filename = "random.txt"
with open(filename, "r") as f:
  header = True
  for line in f:
    if header is False:
      num = line.split(',')
      a.append(num[0])
      b.append(num[1])
      c.append(num[2])
      d.append(num[3])
      e.append(num[4])
    elif header is True:
      header = False
      data = {x.rstrip():0 for x in line.split(',')}
data['methodA'] = a
data['methodB'] = b
data['methodC'] = c
data['methodD'] = d
data['methodE'] = e
for key in data.keys():
  print('\t' + key, end=' ')
print('\n'+' '*6+'-'*75)

for key in data.keys():
  methodX = [float(x) for x in data.get(key)]
  temp = getStats(methodX)
  Max.append(temp[0])
  Min.append(temp[1])
  Mean.append(temp[2])
  Stddev.append(temp[3])
  median.append(temp[4])
  p75.append(temp[5])
  p25.append(temp[6])
stats['Max'] = Max
stats['Min'] = Min
stats['Mean'] = Mean
stats['stddev'] = Stddev
stats['median'] = median
stats['p75'] = p75
stats['p25'] = p25

for key in stats.keys():
  print(key, end='\t  ')
  for value in stats.get(key):
    print("%.2f" % (value), end='\t\t ')
  print(end='\n')

  
```
